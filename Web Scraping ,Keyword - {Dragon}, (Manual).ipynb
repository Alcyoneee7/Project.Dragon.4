```python
#-----------------------------------------     1
#Import File
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import csv
#-----------------------------------------      1


#-----------------------------------------2
#Url , Change Every Projects
#ONLY URL,
driver = webdriver.Chrome()
url= 'https://www.imdb.com/list/ls027224673/?st_dt=&mode=detail&page=1&sort=list_order,asc'
driver.maximize_window()
driver.get(url)

time.sleep(5)
content = driver.page_source.encode('utf-8').strip()
soup = BeautifulSoup(content,"html.parser")


#Change Only (Div,Class)
Container = soup.findAll('div', class_ = "lister-item mode-detail")

#-----------------------------------------2


#-----------------------------------------   4

#Only Run 1 Times, Otherwise Overwrite Things Again & Again
#Csv.file Subject To Change

#with open('Movies 2018-2020.csv', 'w', newline='', encoding='UTF8') as f:
    #writer = csv.writer(f , delimiter ="|")
    #header = ['Title', 'Year', 'Genre',"Avg Ratings", "Votes", "Gross"]
    #writer.writerow(header)

    
#-----------------------------------------  4




#-----------------------------------------            5
#Subject To Change

for p in Container:

    company = p.findAll("h3", class_="lister-item-header")
    for c in company:
        lala = c.find("a").text
    
    skills = p.find("span", class_="lister-item-year text-muted unbold").text
    idont = p.find("span", class_="genre").text.replace(" ","")
    blabla = p.find("span", class_="ipl-rating-star__rating").text if p.find("span", class_="ipl-rating-star__rating") else ""
    nv = p.findAll('span', attrs={'name': 'nv'})
    
  
 #-----------------------------------------          5 
#Ac = Title
#Ad = Year
#Ae = Genre
#Af = Avg Ratings
#bla = Votes
#gro = Grosses

    ac = lala 
    ad  = skills
    ae = idont.strip() 
    
   # af = float(blabla) if len(lala) > 1 else ""
    
    af = float(blabla) if p.find("span", class_="ipl-rating-star__rating") else ""
    bla =  nv[0].text if p.findAll('span', attrs={'name': 'nv'}) else ""
    gro = nv[1].text if len(nv) > 1 else '' 
    
    bliss = [ac,ad,ae,af,bla,gro]    
    
    with open('Movies 2018-2020.csv', 'a', newline='', encoding='UTF8') as f:
        writer = csv.writer(f , delimiter ="|")
        writer.writerow(bliss)
    
driver.quit()

```
